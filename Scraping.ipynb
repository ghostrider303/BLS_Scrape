```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL of the website
url = "https://data.bls.gov/cgi-bin/surveymost?la"

# Fetch the webpage content
response = requests.get(url)
html_content = response.content

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all elements with a specific class
# For example, let's say the class name is "example-class"
class_elements = soup.find('div', class_="verdana md")

# Initialize a list to store all <li> elements
list_elements = class_elements.find('li')

u=[]
s=[]
a_=list_elements.find_all('a')

for t in a_:
    h=t.get('href')
    te=t.text.strip()
    if h:
        u.append(h)
        s.append(te)
        
df= pd.DataFrame({'State': s,'Url':u})
```

```python
import pandas as pd
import csv
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
opts=webdriver.ChromeOptions()
opts.headless=True
driver = webdriver.Firefox(executable_path=r"C:\Users\303an\OneDrive\Documents\geckodriver.exe")
```

```python
url = f'https://data.bls.gov/cgi-bin/surveymost?la+02'
driver.get(url)

def scrape_data(driver):
    try:
        # Define a WebDriverWait object with a maximum timeout of 10 seconds
        # This will be used to wait for elements to appear on the page
        wait = WebDriverWait(driver, 10)

        # Wait until all checkboxes are located on the page
        checkboxes = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "input[type='checkbox']")))

        # Check each checkbox
        for checkbox in checkboxes:
            checkbox.click()

        # Find the button with text "Retrieve Data" and click it
        retrieve_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[value='Retrieve data']")))
        retrieve_button.click()

        # Wait until the new page loads
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "regular-data")))

        # Now you can scrape data from all tables
        tables = driver.find_elements(By.XPATH, "//table[contains(@class, 'regular-data') or contains(@class, 'catalog')]")

        for table in tables:
            # Extracting table ID
            table_id = table.get_attribute("id")
            print("Table ID:", table_id)

            # Scraping data from the table
            rows = table.find_elements(By.TAG_NAME, "tr")
            print("$$$$$$$$$$")
            filename = f"{table_id}.csv"
            with open(filename, "w", newline="", encoding="utf-8") as csvfile:
                writer = csv.writer(csvfile)
                for row in rows:
                    cells = row.find_elements(By.TAG_NAME, "td")
                    row_data = [cell.text for cell in cells]
                    print(row_data)
                    writer.writerow(row_data)

    except Exception as e:
        # If there's an exception, print the error message
        print("An error occurred:", str(e))

    finally:
        # After scraping, close the web driver
        driver.quit()
```

```python
# Main:
scrape_data(driver)
```

```python
def scrape_data_for_urls(driver, urls):
    try:
        for url in urls:
            print(url)
            driver.get(url)

            # Extract folder name from the URL
            folder_name = url[-2:]

            # Create folder if it doesn't exist
            folder_path = os.path.join(os.getcwd(), folder_name)
            if not os.path.exists(folder_path):
                os.makedirs(folder_path)

            # Define a WebDriverWait object with a maximum timeout of 10 seconds
            # This will be used to wait for elements to appear on the page
            wait = WebDriverWait(driver, 2)

            # Wait until all checkboxes are located on the page
            checkboxes = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "input[type='checkbox']")))

            # Check each checkbox
            for checkbox in checkboxes:
                checkbox.click()

            # Find the button with text "Retrieve Data" and click it
            retrieve_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[value='Retrieve data']")))
            retrieve_button.click()

            # Wait until the new page loads
            wait.until(EC.presence_of_element_located((By.CLASS_NAME, "regular-data")))

            # Now you can scrape data from all tables
            tables = driver.find_elements(By.XPATH, "//table[contains(@class, 'regular-data') or contains(@class, 'catalog')]")

            for table in tables:
                # Extracting table ID
                table_id = table.get_attribute("id")

                # Scraping data from the table
                rows = table.find_elements(By.TAG_NAME, "tr")

                # Define file path to save CSV file
                filename = os.path.join(folder_path, f"{table_id}.csv")

                with open(filename, "w", newline="", encoding="utf-8") as csvfile:
                    writer = csv.writer(csvfile)
                    for row in rows:
                        cells = row.find_elements(By.TAG_NAME, "td")
                        row_data = [cell.text for cell in cells]
                        writer.writerow(row_data)

    except Exception as e:
        # If there's an exception, print the error message
        print("An error occurred:", str(e))

    finally:
        # After scraping all URLs, close the web driver
        driver.quit()

# List of URLs
urls = u

# Call the function with the driver and the list of URLs
scrape_data_for_urls(driver, urls)
```

```python
# Function to read data from catalog_id file
def read_catalog_file(catalog_file):
    df = pd.read_csv(catalog_file, header=None)
    # Extract values from 2nd to 5th row
    values = df.iloc[1:5, 0].tolist()
    return values
```

```python
# Function to update corresponding table_id file
def update_table_file(table_file, values):
    df = pd.read_csv(table_file)
    # Delete the last row
    df.drop(df.tail(1).index, inplace=True)
    # Add values as new columns in the corresponding table_id file
    df['Seasonally Adjusted'] = values[0]
    df['Area'] = values[1]
    df['State_or_MSA'] = values[2]
    df['State'] = values[3]
    df.to_csv(table_file, index=False)
```

```python
# Path to the folder containing files
folder_path = './'

# Iterate through each file in the folder
for filename in os.listdir(folder_path):
    if filename.startswith('catalog'):
        print(filename)
        # Extract id from filename
        catalog_id = filename[-5]
        print(catalog_id)
        # Read values from catalog_id file
        values = read_catalog_file(os.path.join(folder_path, filename))
        print(values)
        # Find corresponding table_id file
        table_filename = f'table{catalog_id}.csv'
        print(table_filename)
        table_file_path = os.path.join(folder_path, table_filename)
        if os.path.exists(table_file_path):
            # Update corresponding table_id file
            update_table_file(table_file_path, values)
            print('yes')
```
